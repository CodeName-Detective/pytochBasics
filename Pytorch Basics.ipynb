## Tesnor Operations and Gradients

```
x = torch.tensor(3.)
w = torch.tensor(4.,requires_grad=True)
b = torch.tensor(5.,requires_grad=True)
```

```
y = w * x + b
y
```

- Pytorch is special because it can automatically compute the derivative of y w.r.t. to the tensors that have __required_grad__ set to __True__. To compute the derivatives we can call __.backward__ method on our result __y__.

```
y.backward()
```

```
x.grad #We have not set requires_grad to True #dy/dx
```

```
w.grad # dy/dw
```

```
b.grad # dy/db
```
